```python
# 2.Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 3.Load Dataset
import pandas as pd

df = pd.read_csv("telecom_python_dataset.csv")
df.head()
```

```python
# 4.Basic Dataset Information
df.info()
# Dataset Shape
print(df.shape)
```

```python
# 5.Check Missing Values
df.isnull().sum()
```

```python
# 6.Summary Statistics
df.describe()
```

```python
# 7.KPI Analysis
# Total Customers
total_customers = df.shape[0]

print(total_customers)
# Average ARPU
avg_arpu = df['arpu_8'].mean()

print(avg_arpu)
# Average Recharge Amount
avg_recharge = df['total_rech_amt_8'].mean()

print(avg_recharge)
# Average Recharge Count
avg_recharge_count = df['total_rech_num_8'].mean()

print(avg_recharge_count)
# Total Revenue
total_revenue = df['arpu_8'].sum()

print(total_revenue)
```

```python
# 8.ARPU Distribution Chart
plt.figure(figsize=(8,5))

plt.hist(df['arpu_8'], bins=30)

plt.title("ARPU Distribution")

plt.xlabel("ARPU")

plt.ylabel("Customers")

plt.show()
```

```python
# 9.Recharge Amount Distribution
plt.figure(figsize=(8,5))

plt.hist(df['total_rech_amt_8'], bins=30)

plt.title("Recharge Amount Distribution")

plt.xlabel("Recharge Amount")

plt.ylabel("Customers")

plt.show()
```

```python
# 10.Revenue by ARPU Segment
df['ARPU Segment(8) (new created)'].value_counts()
# Visualization:
df['ARPU Segment(8) (new created)'].value_counts().plot(kind='bar')

plt.title("Customer Distribution by ARPU Segment")

plt.show()
```

```python
# 11.Revenue by Recharge Segment
df.groupby('Recharge_Segment (8)')['arpu_8'].mean()
# Visualization:
df.groupby('Recharge_Segment (8)')['arpu_8'].mean().plot(kind='bar')

plt.title("Average ARPU by Recharge Segment")

plt.show()
```

```python
# 12.Revenue by Usage Segment
df.groupby('Usage_Segment')['arpu_8'].mean()
# Visualization:
df.groupby('Usage_Segment')['arpu_8'].mean().plot(kind='bar')

plt.title("Average ARPU by Usage Segment")

plt.show()
```

```python
# 13.Top 10 Revenue Customers
top10 = df.nlargest(10,'arpu_8')

print(top10[['mobile_number','arpu_8']])
```

```python
# 14.Correlation Analysis
numeric_df = df.select_dtypes(include=np.number)
# Correlation Matrix
corr = numeric_df.corr()

print(corr)
# Heatmap
plt.figure(figsize=(12,8))

sns.heatmap(corr)

plt.title("Correlation Matrix")

plt.show()
```

```python
# 15.Revenue Trend Analysis
trend = {
    'Month':['Month6','Month7','Month8'],
    'ARPU':[
        df['arpu_6'].mean(),
        df['arpu_7'].mean(),
        df['arpu_8'].mean()
    ]
}

trend_df = pd.DataFrame(trend)
# Plot:
plt.figure(figsize=(8,5))

plt.plot(trend_df['Month'],
         trend_df['ARPU'],
         marker='o')

plt.title("Average ARPU Trend")

plt.show()
```

```python
# 16.Recharge Trend Analysis
recharge = {
    'Month':['Month6','Month7','Month8'],
    'Recharge':[
        df['total_rech_amt_6'].mean(),
        df['total_rech_amt_7'].mean(),
        df['total_rech_amt_8'].mean()
    ]
}

recharge_df = pd.DataFrame(recharge)
# Plot:
plt.figure(figsize=(8,5))

plt.plot(recharge_df['Month'],
         recharge_df['Recharge'],
         marker='o')

plt.title("Recharge Trend")

plt.show()
```

```python
# 17. Insights
1.Majority of customers belong to Low Value ARPU segment.
2.High recharge spenders generate highest revenue.
Heavy users contribute greater ARPU.
Recharge frequency positively impacts revenue.
Revenue trend is increasing/decreasing across months.
Data usage strongly correlates with ARPU.
```
